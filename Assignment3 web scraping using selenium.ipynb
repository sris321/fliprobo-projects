#### Ques7. Write a program to scrap all the available details of best gaming laptops from digit.in. 

```python
dr=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

dr.get("https://digit.in./")
time.sleep(5)

search_btn=dr.find_element_by_xpath("//div[@class='search']")
search_btn.click()

search_bar=dr.find_element_by_xpath("//input[@id='globalPageSearchText']")
search_bar.send_keys("gaming laptop")
search_bar.click()

time.sleep(5)
```

```python
title=[]
price=[]
os=[]
display=[]
processor=[]
release_date=[]
rating=[]
```

```python
laptop_url=[]

for i in dr.find_elements_by_xpath("//div[@class='searchPage']/a"):
    laptop_url.append(i.get_attribute('href'))

laptop_url
```

```python
for url in laptop_url:
    dr.get(url)
    tl=dr.find_element_by_xpath("//div[@class='heading-wraper']/h1")
    title.append(tl.text)
    
print(len(title))
```

```python
for url in laptop_url:
    dr.get(url)
    pr=dr.find_element_by_xpath("//div[@class='Block-price']/b")
    price.append(pr.text)
    
print(len(price))
```

```python
for url in laptop_url:
    dr.get(url)
    rd=dr.find_element_by_xpath("/html/body/div[7]/div/div/div[4]/div[1]/b")
    if rd.text is None:
        release_date.append("--")
    else:
        release_date.append(rd.text)
        
print(len(release_date))
```

```python
for url in laptop_url:
    dr.get(url)
    s=dr.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div[1]/div[2]/div[3]/div/ul/li[1]/div")
    if s.text is None:
        os.append("--")
    else:
        os.append(s.text)
        
print(len(os))
```

```python
for url in laptop_url:
    dr.get(url)
    dis=dr.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div[1]/div[2]/div[3]/div/ul/li[2]/div")
    if dis.text is None:
        display.append("--")
    else:
        display.append(dis.text)
        
print(len(display))
```

```python
for url in laptop_url:
    dr.get(url)
    s=dr.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div[1]/div[2]/div[3]/div/ul/li[3]/div")
    if s.text is None:
        processor.append("--")
    else:
        processor.append(s.text)
        
print(len(processor))
```

```python
for url in laptop_url:
    dr.get(url)
    s=dr.find_element_by_xpath("//div[@class='Over-allrating']/div[2]")
    if s.text is None:
        rating.append("--")
    else:
        rating.append(s.text)
        
print(len(rating))
```

```python
img_url=[]

for url in laptop_url:
    dr.get(url)
    ml=dr.find_element_by_xpath("//div[@id='img-01']")
    img_url.append(ml.get_attribute('style'))
    

print(len(img_url))    
```

```python
Gaming_laptop=pd.DataFrame({})
Gaming_laptop['Title']=title
Gaming_laptop['OS']=os
Gaming_laptop['Display']=display
Gaming_laptop['Processor']=processor
Gaming_laptop['Price']=price
Gaming_laptop['Release_Date']=release_date
Gaming_laptop['Rating']=rating
Gaming_laptop['Img_url']=img_url

Gaming_laptop
```

#### Ques10.Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
requests.get("https://www.hostelworld.com/")
```

```python
dr=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

dr.get("https://www.hostelworld.com/")
time.sleep(5)
```

```python
search_bar=dr.find_element_by_xpath("//input[@class='search-input']")
search_bar.send_keys("london")
```

```python
dr.find_element_by_xpath("//li[@tabindex='-1']/div").click()
```

```python
btn=dr.find_element_by_xpath("//button[@id='search-button']")
btn.click()
time.sleep(5)
```

```python
name_urls=[]
for j in range(3):
    for i in dr.find_elements_by_xpath("//h2[@class='title title-6']/a"):
        name_urls.append(i.get_attribute('href'))
        
    try:
        dr.find_element_by_xpath("//div[@class='pagination-item pagination-next']").click()
        time.sleep(3)
    except NoSuchElementException:
        pass
    
print(len(name_urls))
```

```python
name=[]

for ur in name_urls:
    dr.get(ur)
    time.sleep(2)
    tag=dr.find_element_by_xpath("//div[@class='title-2']")
    name.append(tag.text)
            
print(len(name))
```

```python
rat=[]

for ur in name_urls:
    dr.get(ur)
    time.sleep(2)
    try:
        tag=dr.find_element_by_xpath("//div[@class='score orange big']")
        rat.append(tag.text)
    except NoSuchElementException:
        try:
            tag=dr.find_element_by_xpath("//div[@class='score gray big']")
            rat.append(tag.text)

        except NoSuchElementException:
            rat.append("--")
print(len(rat))
```

```python
total_rev=[]

for ur in name_urls:
    dr.get(ur)
    time.sleep(2)
    try:
        tag=dr.find_element_by_xpath("//div[@class='reviews']")
        total_rev.append(tag.text)
    except NoSuchElementException:
        total_rev.append("--")
    
print(len(total_rev))
```

```python
fac=[]
for url in name_urls:
    dr.get(url)
    time.sleep(3)
    tag=dr.find_element_by_xpath("//ul[@class='facility-sections']")
    fac.append(tag.text)
    
print(len(fac))
```

```python
property_desc=[]
for url in name_urls:
    dr.get(url)
    time.sleep(3)
    try:
        dr.find_element_by_xpath("//a[@class='toggle-content']").click()
        time.sleep(2)
        tag=dr.find_element_by_xpath("//div[@style='-webkit-line-clamp:12;']")
        property_desc.append(tag.text)
    except NoSuchElementException:
        tag=dr.find_element_by_xpath("//div[@style='-webkit-line-clamp:12;']")
        property_desc.append(tag.text)
    
print(len(property_desc))
```

```python
distance=[]

for j in dr.find_elements_by_xpath("//span[@class='description']"):
     distance.append(j.text)
        
dr.find_element_by_xpath("//div[@class='pagination-item pagination-next']").click()

for j in dr.find_elements_by_xpath("//span[@class='description']"):
     distance.append(j.text)

dr.find_element_by_xpath("//div[@class='pagination-item pagination-next']").click()

for j in dr.find_elements_by_xpath("//span[@class='description']"):
     distance.append(j.text)
        
    
print(len(distance))
```

```python
price=[]

for j in range(3):
    for i in dr.find_elements_by_xpath("//div[@class='prices-col']"):
        price.append(i.text)
    try:
        dr.find_element_by_xpath("//div[@class='pagination-item pagination-next']").click()
        time.sleep(2)
    except NoSuchElementException:
        pass

print(len(price))
```

#### Ques6.Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

driver.get("https://trak.in/")
time.sleep(5)
```

```python
fund_dl=driver.find_element_by_xpath("//li[@id='menu-item-51510']")
fund_dl.click()
time.sleep(3)
```

```python
date=[]
name=[]
ver=[]
sub_ver=[]
loc=[]
investor_name=[]
ins_type=[]
amt=[]
for j in range(2):
    for i in driver.find_elements_by_xpath("//td[@class='column-2']"):
        date.append(i.text)
    driver.find_element_by_xpath("//a[@class='paginate_button next']").click()

for i in driver.find_elements_by_xpath("//td[@class='column-3']"):
    name.append(i.text)
    
for i in driver.find_elements_by_xpath("//td[@class='column-4']"):
    ver.append(i.text)
    
for i in driver.find_elements_by_xpath("//td[@class='column-5']"):
    sub_ver.append(i.text)
    
for i in driver.find_elements_by_xpath("//td[@class='column-6']"):
    loc.append(i.text)
    
for i in driver.find_elements_by_xpath("//td[@class='column-7']"):
    investor_name.append(i.text)
    
for i in driver.find_elements_by_xpath("//td[@class='column-8']"):
    ins_type.append(i.text)
    
for i in driver.find_elements_by_xpath("//td[@class='column-9']"):
    amt.append(i.text)
    
print(len(date),len(name),len(ver),len(sub_ver),len(loc),len(investor_name),len(ins_type),len(amt))
```

```python
Funding_Deal=pd.DataFrame({})
Funding_Deal['Date']=date
Funding_Deal['Startup_name']=name
Funding_Deal['Vertical']=ver
Funding_Deal['Sub_vertical']=sub_ver
Funding_Deal['Location']=loc
Funding_Deal['Investors_name']=investor_name
Funding_Deal['Invesment_type']=ins_type
Funding_Deal['Amount']=amt

Funding_Deal
```
