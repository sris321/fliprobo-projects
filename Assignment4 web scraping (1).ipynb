### Ques1.. Scrape the details of most viewed videos on YouTube from Wikipedia:Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

#### You need to find following details:
#### A) Rank
#### B) Name
#### C) Artist
#### D) Upload date
#### E) Views

```python
requests.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(5)
```

```python
name=[]

try:

    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")

    for i in names:

        name.append(i.text)

except NoSuchElementException:

    name.append('No details available')

except StaleElementReferenceException:

    name.append('No details available')
    
print(len(name))
```

```python
rank=[]

try:
    rk=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in rk:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('No details available')
        
print(len(rank))
```

```python
artist=[]

try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        artist.append(i.text)
        
except NoSuchElementException:
    artist.append("No details available")
    
print(len(artist))
```

```python
date=[]

try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        date.append(i.text)
except NoSuchElementException:
    date.append("No details available")
    
print(len(date))
```

```python
view=[]
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        view.append(i.text)
except NoSuchElementException:
    view.append("No details available")
    
print(len(view))
```

```python
Youtube_mostviewed_video=pd.DataFrame({})
Youtube_mostviewed_video['Rank']=rank
Youtube_mostviewed_video['Name']=name
Youtube_mostviewed_video['Artist']=artist
Youtube_mostviewed_video['Upload date']=date
Youtube_mostviewed_video['Views']=view

Youtube_mostviewed_video
```

```python
requests.get("https://github.com/")
```

```python
driver=webdriver.Chrome("chromedriver.exe")

driver.get("https://github.com/")
```

```python
explore_btn=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
explore_btn.click()
time.sleep(3)
```

```python
trending_btn=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
trending_btn.click()
time.sleep(3)
```

```python
title=[]

for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']"):
    title.append(i.text)
    
print(len(title))
```

```python
url=[]
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    url.append(i.get_attribute("href"))
print(len(url))
```

```python
desc=[]
for ur in url:
    driver.get(ur)
    time.sleep(3)
    try:
        tag=driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        desc.append(tag.text)
    except NoSuchElementException:
        desc.append("not available")
    
print(len(desc))
```

```python
count=[]
for ur in url:
    driver.get(ur)
    time.sleep(3)
    try:
        tag=driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[5]/div/h2/a/span")
        count.append(tag.text)
    except NoSuchElementException:
        try:
            tag=driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[4]/div/h2/a/span")
            count.append(tag.text)
        except NoSuchElementException:
            try:
                tag=driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[6]/div/h2/a/span")
                count.append(tag.text)
            except NoSuchElementException:
                count.append("--")
    
print(len(count))
```

```python
lang=[]
for ur in url:
    driver.get(ur)
    time.sleep(3)
    try:
        tag=driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[6]/div/ul")
        lang.append(tag.text)
    except NoSuchElementException:
        try:
            tag=driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[5]/div/ul")
            lang.append(tag.text)
        except NoSuchElementException:
            lang.append("--")
    
print(len(lang))
```

```python
Github_repository=pd.DataFrame({})
Github_repository['Repository_title']=title
Github_repository['Repository_desc']=desc
Github_repository['Contributors_count']=count
Github_repository['Language_used']=lang

Github_repository
```

```python
Github_repository['Contributors_count']=Github_repository['Contributors_count'].replace('','49')
```

```python
Github_repository['Language_used']=Github_repository['Language_used'].replace('','--')
```
